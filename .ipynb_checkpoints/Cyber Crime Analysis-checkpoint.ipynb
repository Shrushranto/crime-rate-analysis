```python
def getCat(data):
    category = []
    for i in data:
        if i in Crime2018["Crime Head"].values:
            getRow = Crime2018[Crime2018["Crime Head"] == i]
            category_values = list(set(getRow["Category (Col.2)"].values))
        else:
            category_values = ["Not Found"]  
        category.append(", ".join(category_values))  
    return category
```

### Checking the Trend of Cyber Crime

```python
# Trend in reported cases
reported_trend = finalCyber.groupby('year')['Cases Reported during the year'].sum().reset_index()
reported_trend['year'] = reported_trend['year'].astype(int) 

plt.figure(figsize=(8,5))
sns.lineplot(data=reported_trend, x='year', y='Cases Reported during the year', marker='o')
plt.title("Trend in Reported Cybercrime Cases (2018–2021)")
plt.xlabel("Year")
plt.ylabel("Number of Reported Cases")
plt.xticks(reported_trend['year'])  
plt.grid(True)
plt.show()
```

### Top 3 crime heads per year

```python
top_3 = finalCyber.groupby(['year', 'Crime Head'])['Cases Reported during the year'].sum().reset_index()
top_3_sorted = top_3.sort_values(['year', 'Cases Reported during the year'], ascending=[True, False])
top_3_each_year = top_3_sorted.groupby('year').head(3)

plt.figure(figsize=(10,6))
sns.barplot(data=top_3_each_year, x='year', y='Cases Reported during the year', hue='Crime Head')
plt.title("Top 3 Crime Heads per Year")
plt.ylabel("Cases Reported")
plt.legend(title="Crime Head", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

### Top 3 Reported Cybercrime Categories (2018–2021) – Insights
<b>2018</b>

<b>Most reported</b>: Identity Theft

Followed by: Offences Other Than Ransomware, Cheating by Personation
→ Focus on personal data misuse and general cyber offences dominated this year.

<b>2019</b>

<b>Most reported</b>: Identity Theft

Followed by: Cheating by Personation, Offences Other Than Ransomware
→ Continuation of identity-related crimes, with impersonation-based frauds rising.

<b>2020</b>

<b>Most reported</b>: Cheating by Personation

Followed by: Identity Theft, Cheating
→ A shift toward social engineering scams, likely influenced by increased online activity during the pandemic.

<b>2021</b>

<b>Most reported</b>: Cheating by Personation

Followed by: Cheating, Online Banking Fraud
→ Financially motivated cybercrimes became more prominent, reflecting a rise in online transactions and banking frauds.

<b>Key Takeaways</b>
<li>Identity Theft was the leading crime in 2018 and 2019, highlighting concerns around data privacy and misuse of personal information.</li>

<li>Cheating by Personation emerged as the top threat in 2020 and 2021, showing how impersonation tactics have become increasingly sophisticated.</li>

<li>Online Banking Fraud appeared in the top 3 in 2021, indicating a shift toward more targeted financial cybercrimes.</li>



### Why there were no cases for Identity theft as the year passes by

```python
# Filter for 'Identity Theft' cases
identity_theft_trend = finalCyber[finalCyber['Crime Head'] == 'Identity Theft']

# Group by year: Reported and Pending
identity_reported = identity_theft_trend.groupby('year')['Cases Reported during the year'].sum().reset_index()
identity_pending = identity_theft_trend.groupby('year')['Cases Pending Investigation at End of the Year'].sum().reset_index()
merged_identity = identity_reported.merge(identity_pending, on='year')
merged_identity.columns = ['year', 'Reported', 'Pending']

plt.figure(figsize=(8, 5))
merged_identity.set_index('year').plot(kind='bar', figsize=(8, 5))
plt.title("Identity Theft: Reported vs Pending Cases (2018–2021)")
plt.ylabel("Number of Cases")
plt.xlabel("Year")
plt.grid(True)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Compare Identity Theft vs other rising crime types
top_crimes = ['Identity Theft', 'Cheating by Personation', 'Cheating']
trend_top = finalCyber[finalCyber['Crime Head'].isin(top_crimes)]
trend_grouped = trend_top.groupby(['year', 'Crime Head'])['Cases Reported during the year'].sum().reset_index()

# Plot comparison trend
plt.figure(figsize=(9, 5))
sns.lineplot(data=trend_grouped, x='year', y='Cases Reported during the year', hue='Crime Head', marker='o')
plt.title("Trend of Top Cybercrimes (2018–2021)")
plt.ylabel("Reported Cases")
plt.xlabel("Year")
plt.grid(True)
plt.xticks(trend_grouped['year'].unique())
plt.tight_layout()
plt.show()
```

```python
# if the cases were closed what was the reason behind
```

```python
# Checking closure reasons for Identity Theft
final_report_cols = [
    'Final Report - Cases Ended as Final Report False',
    'Final Report - Cases Ended as Mistake of Fact or of Law or Civil Dispute',
    'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue'
]

identity_closure = identity_theft_trend.groupby('year')[final_report_cols].sum().reset_index()

identity_closure.set_index('year').plot(kind='bar', stacked=True, figsize=(9, 5))
plt.title("Identity Theft: Reasons for Case Closure (2018–2021)")
plt.ylabel("Number of Cases")
plt.xlabel("Year")
plt.legend(title="Closure Reason", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

identity_closure.head()  
```

## Which cybercrime categories show the highest number of cases reported during the years 2018-2021?

```python
# Filter for years 2018 to 2021
cyber_df = finalCyber[finalCyber['year'].between(2018, 2021)]
category_reported = cyber_df.groupby(['year', 'Category'])['Cases Reported during the year'].sum().reset_index()

# visualization
plt.figure(figsize=(12,6))
sns.barplot(data=category_reported, x='year', y='Cases Reported during the year', hue='Category')
plt.title("Cybercrime Cases Reported per Category (2018–2021)")
plt.ylabel("Cases Reported")
plt.xticks(rotation=45)
plt.legend(title="Category", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

## What is the percentage of cybercrime cases that remain pending at the end of the year compared to the total number of cases for investigation?

```python
# Calculate Pending Percentage
cyber_df['Pending Percentage'] = (
    cyber_df['Cases Pending Investigation at End of the Year'] / 
    cyber_df['Total Cases for Investigation']
) * 100

# Group by year and compute average
pendency_trend = cyber_df.groupby('year')['Pending Percentage'].mean().reset_index()
pendency_trend['year'] = pendency_trend['year'].astype(int)

# Plot
plt.figure(figsize=(8, 5))
sns.lineplot(data=pendency_trend, x='year', y='Pending Percentage', marker='o')
plt.title("Average Pendency Percentage of Cybercrime Cases (2018–2021)")
plt.ylabel("Pendency (%)")
plt.xlabel("Year")
plt.grid(True)

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.show()

```

### As there is a major hike in the year 2020, now let us find that which are the top crime heads affecting to this

```python
# Filter for 2020
data_2020 = cyber_df[cyber_df['year'] == 2020]
top_pending_2020 = data_2020.sort_values(
    by='Cases Pending Investigation at End of the Year', 
    ascending=False
)[['Crime Head', 'Cases Pending Investigation at End of the Year']].head(5)

print("Top 5 Crime Heads with Highest Pending Cases in 2020:")
top_pending_2020.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=top_pending_2020, 
            x='Cases Pending Investigation at End of the Year', 
            y='Crime Head', 
            palette='coolwarm')

plt.title("Top 5 Crime Heads with Highest Pending Cases in 2020")
plt.xlabel("Pending Cases")
plt.ylabel("Crime Head")
plt.tight_layout()
plt.show()
```

## What is the disposition rate (cases disposed off) for cybercrimes across the years?


```python
# Calculating the  Disposition Rate
cyber_df['Disposition Rate'] = (
    cyber_df['Total Cases Disposed Off by Police'] / 
    cyber_df['Total Cases for Investigation']
) * 100
```

```python
# Group by year and compute average
disp_trend = cyber_df.groupby('year')['Disposition Rate'].mean().reset_index()

disp_trend['year'] = disp_trend['year'].astype(int)

# Plot
plt.figure(figsize=(8,5))
sns.lineplot(data=disp_trend, x='year', y='Disposition Rate', marker='o', color='green')
plt.title("Disposition Rate for Cybercrime Cases (2018–2021)")
plt.ylabel("Disposition Rate (%)")
plt.xlabel("Year")
plt.grid(True)

plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
plt.show()
```

```python
# Filter data for the year 2021
data_2021 = cyber_df[cyber_df['year'] == 2021]

# Group by 'Category' and 'Crime Head' to sum the number of disposed cases
crime_disposed_2021 = data_2021.groupby(['Category', 'Crime Head'])['Total Cases Disposed Off by Police'].sum().reset_index()

# sorting in ascending order
sorted_disposed_2021 = crime_disposed_2021.sort_values(
    by='Total Cases Disposed Off by Police', 
    ascending=False
)

top_5_disposed_2021 = sorted_disposed_2021.groupby('Category').head(5)

print("Top 5 Disposed Crimes in Each Category (2021):")
top_5_disposed_2021.head()
```

```python
# Plot the top 5 crimes for each category
plt.figure(figsize=(12, 8))

sns.barplot(data=top_5_disposed_2021, 
            x='Crime Head', 
            y='Total Cases Disposed Off by Police', 
            hue='Category', 
            palette='Set2')

# Title and labels
plt.title("Top 5 Disposed Crimes by Category (2021)")
plt.ylabel("Number of Cases Disposed")
plt.xlabel("Crime Head")
plt.xticks(rotation=45, ha='right')

# Display grid and legend
plt.grid(True)
plt.legend(title="Category")

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.show()
```

## What are the reasons for the closure of cybercrime cases (e.g., cases ended as false, insufficient evidence, mistake of fact)?



```python
closure_cols = [
    'Final Report - Cases Ended as FR Non Cognizable',
    'Final Report - Cases Ended as Final Report False',
    'Final Report - Cases Ended as Mistake of Fact or of Law or Civil Dispute',
    'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue',
    'Final Report - Cases Abated during Investigation'
]

closure_summary = cyber_df[closure_cols].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=closure_summary.values, y=closure_summary.index, palette='Reds_r')
plt.title("Reasons for Closure of Cybercrime Cases (2018–2021)")
plt.xlabel("Number of Cases Closed")
plt.tight_layout()
plt.show()
```

```python
# Filter the data for cases where they were closed due to "Insufficient Evidence or Untraced or No Clue"
insufficient_evidence_cases = cyber_df[['Category', 'Crime Head', 
                                       'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue']]

crime_head_cases_closed = insufficient_evidence_cases.groupby('Crime Head')[
    'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue'
].sum().reset_index()

# Sort values to get the top crime heads with the most cases closed
top_crime_heads_closed = crime_head_cases_closed.sort_values(
    by='Final Report - Cases True but Insufficient Evidence or Untraced or No Clue', 
    ascending=False
).head(10)

# Plot
plt.figure(figsize=(10,6))
sns.barplot(data=top_crime_heads_closed, 
            x='Crime Head', 
            y='Final Report - Cases True but Insufficient Evidence or Untraced or No Clue', 
            palette='viridis')

plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.2)

plt.title("Top 10 Crime Heads with Highest Cases Closed Due to Insufficient Evidence or No Clue (2018–2021)")
plt.ylabel("Number of Cases Closed")
plt.xlabel("Crime Head")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```
