```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
```

```python
df18=pd.read_csv("IPC_18.csv")
```

```python
df19=pd.read_csv("IPC_19.csv")
```

```python
df20=pd.read_csv("IPC_20.csv")
```

```python
df21=pd.read_csv("IPC_21.csv")
```

```python
# Adding category column to all tables

for df in [df19, df20, df21]:
    df.insert(1, 'Category', '')

def getCat(data):
    category = []
    for i in data:
        if i in df18["Crime Head"].values:
            getRow = df18[df18["Crime Head"] == i]
            category_values = list(set(getRow["Category"].values))
        else:
            category_values = ["Not Found"]
        category.append(", ".join(category_values))
    return category  # <-- moved outside the for loop

df19["Category"] = getCat(df19["Crime Head"])
df20["Category"] = getCat(df20["Crime Head"])
df21["Category"] = getCat(df21["Crime Head"])
```

```python
import re

def clean_column_names(df):
    df.columns = [re.sub(r'\s*\(.*?\)', '', col).strip() for col in df.columns]
    return df

# Apply to all DataFrames
df18 = clean_column_names(df18)
df19 = clean_column_names(df19)
df20 = clean_column_names(df20)
df21 = clean_column_names(df21)
```

```python
df20.columns = df20.columns.str.replace(r'\s+', '-', regex=True)
```

```python
df21.columns = df21.columns.str.replace(r' - Col\.$', '', regex=True)
df21.columns = df21.columns.str.replace(r' Col\.$', '', regex=True)
```

# 1. How have the number of IPC crime cases reported and investigated evolved over the years?

```python
import matplotlib.pyplot as plt

# Group by Year and sum up the relevant columns
trend_data = merged_df.groupby('Year')[[
    'Cases Reported during the year',
    'Total Cases for Investigation'
]].sum().reset_index()

# Convert values to per 1000 scale
trend_data['Cases Reported during the year'] /= 1000
trend_data['Total Cases for Investigation'] /= 1000

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(trend_data['Year'], trend_data['Cases Reported during the year'], marker='o', label='Cases Reported')
plt.plot(trend_data['Year'], trend_data['Total Cases for Investigation'], marker='s', label='Total Cases for Investigation')

plt.title('IPC Crime Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Cases(in 1000s)')
plt.xticks([2018, 2019, 2020, 2021])  # Ensuring specific years are shown
plt.ylim(0, )                      # Y-axis range as 0 to 100
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

📊 Insights from the Chart:

🔹 Trend of Crime Reporting:

    -If the line for "Cases Reported during the year" shows an upward trend, it suggests an increase in IPC crime
    cases being reported each year.

    -A dip in 2020 might reflect the impact of COVID-19 lockdowns, where public movement was restricted and 
    reporting might have decreased temporarily.

🔹 Trend of Investigations:
    
    -The "Total Cases for Investigation" line may follow a similar pattern, but slightly higher, as it also 
    includes cases carried forward from the previous year.

    -A growing gap between reported and investigated cases over the years might indicate increasing case backlog.

# 2. Which IPC crime categories (e.g., theft, assault, murder) show the highest number of cases reported and pending investigations?

```python
category_summary = merged_df.groupby('Category')[[
    'Cases Reported during the year',
    'Cases Pending Investigation from Previous Year'
]].sum().sort_values(by='Cases Reported during the year', ascending=False)

# Convert values to per 1000 scale
category_summary['Cases Reported during the year']/=1000
category_summary['Cases Pending Investigation from Previous Year']/=1000

# Prepare data
categories = category_summary.index
reported = category_summary['Cases Reported during the year']
pending = category_summary['Cases Pending Investigation from Previous Year']

# Set position of bars on X axis
x = np.arange(len(categories))
width = 0.4  # width of the bars

# Plotting
plt.figure(figsize=(16, 8))
plt.bar(x - width/2, reported, width, label='Reported Cases')
plt.bar(x + width/2, pending, width, label='Pending Cases')

# Labels and titles
plt.xlabel('Category')
plt.ylabel('Number of Cases(in 1000s)')
plt.title('Reported vs Pending Cases by Category')
plt.xticks(ticks=x, labels=categories, rotation=90)
plt.legend()
plt.tight_layout()

plt.show()
```

📊 Insights from the Chart:

🔹 Category-wise Crime Reporting Trend

    -Categories with very high "Cases Reported during the year" (as seen in the top chart) indicate areas where crimes are either more prevalent or more frequently reported.

    -This could reflect increased public awareness, improved access to reporting mechanisms, or a genuine rise in criminal activity in those areas.

    -The large number of reported cases in certain categories might also be tied to digitally traceable crimes or socially sensitive issues where reporting is more encouraged.

🔹 Pending Investigations Reveal System Load

    -Categories with high "Cases Pending Investigation from Previous Year" suggest areas where investigative or judicial bottlenecks may exist.

    -A large number of pending cases can lead to delayed justice, potentially eroding public trust in the system.

    -These categories may require resource allocation, faster case disposal mechanisms, or legal reforms to handle complexity or volume.

🔹 Gap Between Reported and Pending Cases

    -When a category has high reporting but low pending cases, it indicates efficient follow-up and resolution — possibly a model for other categories.

    -Conversely, if pending cases outnumber or closely match reported ones, it could indicate year-over-year accumulation, suggesting an increasing backlog.

    -Monitoring this gap helps understand whether the system is keeping up with the pace of crime reporting or falling behind.

# 2. What is the chargesheeting rate for IPC crimes from 2018 to 2021?

```python
chargesheeting_rate = merged_df.groupby('Year')[
    ['Chargesheeting Rate *100']
].mean().reset_index()
```

```python
chargesheeting_rate = merged_df.groupby('Year')[['Chargesheeting Rate *100']].mean().reset_index()
chargesheeting_rate.columns = ['Year', 'Average Chargesheeting Rate (%)']

# Plotting the chargesheeting rate over the years
plt.figure(figsize=(10, 6))
plt.plot(chargesheeting_rate['Year'], chargesheeting_rate['Average Chargesheeting Rate (%)'],
         marker='o', linestyle='-', color='mediumseagreen')

# Explicitly set x-axis ticks
plt.xticks([2018, 2019, 2020, 2021])

plt.title('Average Chargesheeting Rate for IPC Crimes (2018–2021)')
plt.xlabel('Year')
plt.ylabel('Chargesheeting Rate (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

📊 Insights from the Chart:
    
🔹 Chargesheeting Efficiency Over Time

    -If the line for “Average Chargesheeting Rate (%)” shows a consistent or upward trend between 2018 and 2021, it suggests that law enforcement agencies have become more efficient in completing investigations and filing chargesheets for IPC crimes.

🔹 Possible Impact of COVID-19 (2020)

    -A dip in 2020 may indicate the effect of COVID-19 lockdowns, which likely disrupted police investigations, court functioning, and administrative processes, resulting in delayed or fewer chargesheets being filed that year.

🔹 Inter-Year Comparison

    -A consistent or rising trend from 2018 to 2021 would highlight steady improvements in the justice system’s processing capability, possibly due to digitization, legal reforms, or policy-driven performance metrics at state or national levels.

# 4. What are the percentages of cases withdrawn by the government during the investigation process?


```python
withdrawn_stats = merged_df.groupby('Year')[
    [ 'Total Cases for Investigation','Cases Withdrawn by the Govt during Investigation']
].sum().reset_index()

# Calculate percentage
withdrawn_stats['% Cases Withdrawn'] = (
    withdrawn_stats['Cases Withdrawn by the Govt during Investigation'] /
    withdrawn_stats['Total Cases for Investigation']
) * 100

# Round the numbers for better presentation
withdrawn_stats['Total Cases for Investigation'] = withdrawn_stats['Total Cases for Investigation'].astype(int)
withdrawn_stats['Cases Withdrawn by the Govt during Investigation'] = withdrawn_stats['Cases Withdrawn by the Govt during Investigation'].astype(int)
withdrawn_stats['% Cases Withdrawn'] = withdrawn_stats['% Cases Withdrawn'].round(4)

# Rename columns for readability
withdrawn_stats.columns = [
    'Year',
    'Total Cases for Investigation',
    'Cases Withdrawn by Govt',
    '% Cases Withdrawn'
]

withdrawn_stats.head()
```



📊Insights from the table:

🔹Overall Low Withdrawal Rate

    -Across all four years, less than 0.05% of total IPC cases were withdrawn by the government during the investigation stage.

    -This suggests that withdrawal is a rare outcome and most cases proceed through other channels (chargesheet, closure reports, pending).

🔹Significant Spike in 2021

    -2021 saw a dramatic increase in the number of withdrawn cases (1,747), over 16 times more than in 2020.

    -This caused the withdrawal rate to jump from 0.0019% (2020) to 0.0301% (2021).

    -Possible interpretations:

        Change in government policy or priorities (e.g., backlog clearance).

        COVID-19 pandemic aftermath might have led to dropping non-priority cases.

        Could indicate increased use of administrative discretion in specific states or categories of cases.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace with the actual column names from your merged_df DataFrame
withdrawn_by_govt_column = 'Cases Withdrawn by the Govt during Investigation'
mistake_of_fact_column = 'Final Report - Cases Ended as Mistake of Fact or of Law or Civil Dispute'
insufficient_evidence_column = 'Final Report - Cases True but Insufficient Evidence or Untraced or No Clue'
total_cases_column = 'Total Cases for Investigation'  # Replace with the column representing total cases

# Aggregate the number of cases withdrawn by the government
withdrawn_by_govt = merged_df[withdrawn_by_govt_column].sum()

# Aggregate the number of cases withdrawn due to each reason
withdrawn_due_to_mistake = merged_df[mistake_of_fact_column].sum()
withdrawn_due_to_insufficient_evidence = merged_df[insufficient_evidence_column].sum()

# Calculate the remaining withdrawn cases (total withdrawn - specific reasons)
# remaining_withdrawn_cases = withdrawn_by_govt - (withdrawn_due_to_mistake + withdrawn_due_to_insufficient_evidence)

# Calculate the total number of cases (adjust if necessary)
total_cases = merged_df[total_cases_column].sum()

# Calculate remaining cases not withdrawn by the government
remaining_not_withdrawn_cases = total_cases - withdrawn_by_govt
# Data for the pie chart
data = [
    withdrawn_due_to_mistake,
    withdrawn_due_to_insufficient_evidence,
    remaining_not_withdrawn_cases
]

labels = [
    'Withdrawn: Mistake of Fact / Civil Dispute',
    'Withdrawn: Insufficient Evidence / Untraced',
    'Not Withdrawn by Govt.'
]

colors = ['#b2df8a', '#fdbf6f', '#1f78b4']

# Create the pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    data,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    wedgeprops=dict(edgecolor='white'),
    pctdistance=0.8,
    explode=(0.05, 0.05, 0.05)
)

# Styling for text inside the pie
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(12)

# Add a side legend
plt.legend(wedges, labels, title="Category", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.title("Distribution of Cases Withdrawn and Not Withdrawn by Govt.", fontsize=14)
plt.tight_layout()
plt.show()
```
