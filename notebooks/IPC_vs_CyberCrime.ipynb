# Data Analysis

### How do the trends in cybercrime cases reported compare to those in IPC crimes from 2018-2021?

```python
# Group by year and sum the cases reported for both Cybercrime and IPC crimes
cyber_reported = Cyber.groupby('year')['Cases Reported during the year'].sum().reset_index()
ipc_reported = ipc.groupby('Year')['Cases Reported during the year'].sum().reset_index()
ipc_reported['Cases Reported during the year']/=1000
# Create subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# # Plot trend for Cybercrime cases
sns.lineplot(data=cyber_reported, x='year', y='Cases Reported during the year', marker='o', color='blue', ax=axes[0])
axes[0].set_title("Cybercrime Reported (2018–2021)")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Cases Reported")
axes[0].grid(True)
axes[0].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Plot trend for IPC crimes
sns.lineplot(data=ipc_reported, x='Year', y='Cases Reported during the year', marker='o', color='red', ax=axes[1])
axes[1].set_title("IPC Crimes Reported (2018–2021)")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Cases Reported")
axes[1].grid(True)
axes[1].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()

plt.show()
```

### Which Types of Crimes (Cyber or IPC) Show a Higher Rate of Pending Investigations at the End of the Year?

```python
# Group by year and sum the cases reported for both Cybercrime and IPC crimes
cyber_reported = Cyber.groupby('year')['Cases Pending Investigation at End of the Year'].sum().reset_index()
ipc_reported = ipc.groupby('Year')['Cases Pending Investigation at End of the Year'].sum().reset_index()
ipc_reported['Cases Pending Investigation at End of the Year']/=1000
# Create subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot trend for Cybercrime cases
sns.lineplot(data=cyber_reported, x='year', y='Cases Pending Investigation at End of the Year', marker='o', color='blue', ax=axes[0])
axes[0].set_title("Cybercrime Pending Investigation (2018–2021)")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Cases Pending")
axes[0].grid(True)
axes[0].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

# Plot trend for IPC crimes
sns.lineplot(data=ipc_reported, x='Year', y='Cases Pending Investigation at End of the Year', marker='o', color='red', ax=axes[1])
axes[1].set_title("IPC Crimes Pending Investigation (2018–2021)")
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Cases Pending")
axes[1].grid(True)
axes[1].xaxis.set_major_locator(ticker.MaxNLocator(integer=True))

plt.tight_layout()

plt.show()
```

### What is the Chargesheeting Rate for Cybercrime vs IPC Crime Cases Over the Years?

```python
# Group by Year and calculate the average Chargesheeting Rate for both datasets
cyber_chargesheet_rate = Cyber.groupby('year')['Chargesheeting Rate *100'].mean().reset_index()
ipc_chargesheet_rate = ipc.groupby('Year')['Chargesheeting Rate *100'].mean().reset_index()

merged_chargesheet_rate = pd.merge(cyber_chargesheet_rate, ipc_chargesheet_rate, left_on='year', right_on='Year', suffixes=('_cyber', '_ipc'))

plt.figure(figsize=(10, 6))
width = 0.35 
plt.bar(merged_chargesheet_rate['year'] - width/2, merged_chargesheet_rate['Chargesheeting Rate *100_cyber'], width=width, label='Cybercrime', color='blue')
plt.bar(merged_chargesheet_rate['year'] + width/2, merged_chargesheet_rate['Chargesheeting Rate *100_ipc'], width=width, label='IPC Crime', color='red')

# Add title and labels
plt.title("Chargesheeting Rate Comparison (Cybercrime vs IPC Crimes)")
plt.xlabel("Year")
plt.ylabel("Chargesheeting Rate (%)")
plt.legend()
plt.grid(True)
plt.xticks(merged_chargesheet_rate['year']) 

plt.show()
```

## Insights based on Chargesheeting Rate Comparison (Cybercrimes vs IPC Crimes)
<b>Observation</b>: The chargesheeting rate for IPC crimes consistently outperformed cybercrimes across all years (2018–2021). While IPC crimes maintained a rate above 70%, cybercrimes lingered below 50%.

<b>Insight</b>: This disparity highlights challenges in investigating and filing chargesheets for cybercrimes, possibly due to the complexity of digital evidence and tracing perpetrators.

###  Which Type of Crime (Cybercrime or IPC Crime) Has a Higher Rate of Cases Transferred to Other Agencies or States?

```python
# Group by year and sum the number of cases transferred to other agencies or states
cyber_transferred = Cyber.groupby('year')['Cases Transferred to other State or Agency'].sum().reset_index()
ipc_transferred = ipc.groupby('Year')['Cases Transferred to other State or Agency'].sum().reset_index()

# Plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot for Cybercrime on the first y-axis (left)
sns.lineplot(data=cyber_transferred, x='year', y='Cases Transferred to other State or Agency', marker='o', ax=ax1, color='blue')
ax1.set_xlabel("Year")
ax1.set_ylabel("Cybercrime Cases Transferred", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Create the second y-axis (right) for IPC Crimes
ax2 = ax1.twinx()
sns.lineplot(data=ipc_transferred, x='Year', y='Cases Transferred to other State or Agency', marker='o', ax=ax2, color='red')
ax2.set_ylabel("IPC Crime Cases Transferred", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and final adjustments
plt.title("Comparison of Cases Transferred to Other Agencies or States (Cybercrime vs IPC Crimes)")
plt.gca().xaxis.set_major_locator(ticker.MaxNLocator(integer=True))  
plt.show()
```

## Correlation Matrix
Pearson's Correlation: This method measures the linear relationship between two datasets. If the correlation is close to -1, it indicates a negative correlation (as one increases, the other decreases).<br>
<b>Interpretation of Correlation Coefficient:</b>
<li> -1:  Strong negative correlation (as cybercrimes increase, IPC crimes decrease).</li>

<li>  0:  No correlation (no relationship between the increase in cybercrimes and IPC crimes).</li>

<li> +1:  Strong positive correlation (both increase together).</li>



```python
# Group by year and sum the cases reported for both datasets
cyber_total_reported = Cyber.groupby('year')['Cases Reported during the year'].sum().reset_index()
ipc_total_reported = ipc.groupby('Year')['Cases Reported during the year'].sum().reset_index()

merged_data = pd.merge(cyber_total_reported, ipc_total_reported, left_on='year', right_on='Year', suffixes=('_cyber', '_ipc'))

# Calculating Pearson's correlation between Cybercrime and IPC Crimes
correlation = merged_data['Cases Reported during the year_cyber'].corr(merged_data['Cases Reported during the year_ipc'])

print(f"Pearson's Correlation between Cybercrimes and IPC Crimes: {correlation:.2f}")
```

## Correlation Insight: Cybercrime vs IPC Crime (2018–2021)

<b>Correlation Coefficient:</b> 0.68

<b>Interpretation:</b>
<ul>
  <li><b>Positive Correlation:</b> Both cybercrimes and IPC crimes tend to increase together. An increase in cybercrime often coincides with an increase in IPC crime.</li>
  <li><b>Moderate Strength:</b> The correlation of 0.68 is moderate—not strong enough to suggest a direct cause, but notable enough to indicate a relationship.</li>
</ul>

<b>Possible Explanations:</b>
<ul>
  <li><b>Increase in Online & Offline Crimes:</b> As society becomes more tech-driven, both types of crimes are growing in parallel.</li>
  <li><b>Resource Allocation:</b> With more focus on tackling cybercrime, traditional crime response strategies may be impacted.</li>
  <li><b>Improved Awareness & Reporting:</b> A rise in cybercrime awareness may also lead to increased reporting of other (IPC) crimes.</li>
</ul>

<b>Note:</b> Correlation does not imply causation. Additional analysis is needed to determine any direct impact cybercrime may have on IPC crime rates.
